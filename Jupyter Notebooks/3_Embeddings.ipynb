```python
def embed(data_to_embed, key_to_embed, primary_key, batch_size, verbose=False):

    # Process all texts as batches
    embeddings = model.encode(
        data_to_embed[key_to_embed].tolist(),
        batch_size = batch_size,
        show_progress_bar=verbose
    )

    # Convert embeddings to a DataFrame
    embedded_df = pd.DataFrame(
        embeddings,
        columns=[f'Q{i+1}' for i in range(embeddings.shape[1])]
    )

    # Add the primary key column
    embedded_df.insert(0, primary_key, data_to_embed[primary_key].values)

    return embedded_df
```

```python
def get_batch(df, data_divisions, division):
    """
    Returns the corresponding batch of the DataFrame.

    Args:
        df (pd.DataFrame): The original DataFrame.
        data_divisions (int): Total number of divisions.
        division (int): Number of the division you want (1-indexed).

    Returns:
        pd.DataFrame: The corresponding batch.
    """
    assert 1 <= division <= data_divisions, "division out of range"
    total_rows = len(df)
    batch_size = total_rows // data_divisions
    start_idx = (division - 1) * batch_size
    # The last division takes the remainder (in case it is not exactly divisible)
    if division == data_divisions:
        end_idx = total_rows
    else:
        end_idx = start_idx + batch_size
    return df.iloc[start_idx:end_idx]
```
