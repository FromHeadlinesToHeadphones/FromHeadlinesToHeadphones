After that, we need to prepare the columns of both dataframes to be able to merge them and find the song_id. For achieving this, we will convert everything to lowercase and strip spaces, to try to have as least differences in format as possible.

```python
# Create combined key in hits_df
hits_df['match_key'] = (
    hits_df['Title'].str.lower().str.strip() + "___" + hits_df['Artist'].str.lower().str.strip()
)

# Create combined key in songs_df
songs_df['match_key'] = (
    songs_df['song'].str.lower().str.strip() + "___" + songs_df['Artists'].str.lower().str.strip()
)
```

```python
# Keep only the first song_id per match_key
songs_dedup = songs_df.reset_index().drop_duplicates(subset='match_key', keep='first')[['song_id', 'match_key']]

# Merge hits_df with deduplicated song_id table
hits_df_indexed = hits_df.merge(
    songs_dedup,
    on='match_key',
    how='left',
    validate='many_to_one'  # Ensures that each match_key in hits_df maps to at most one song_id
)
```

```python
# Count matched rows
num_matched = hits_df_indexed['song_id'].notna().sum()

# Count total rows
total_rows = len(hits_df_indexed)

# Calculate percentage
percentage_matched = (num_matched / total_rows) * 100

print(f"Matched songs: {num_matched} of {total_rows} ({percentage_matched:.2f}%)")
```

```python
unmatched_df = hits_df_indexed.loc[hits_df_indexed['song_id'].isna()]
unmatched_df
```

```python
# Get unmatched entries
unmatched_df = hits_df_indexed.loc[hits_df_indexed['song_id'].isna() & hits_df_indexed['match_key'].notna()].copy()

# Prepare reference keys
songs_meta_df = songs_df.reset_index()[['song_id', 'match_key', 'song', 'Artists']]
valid_keys = songs_meta_df['match_key'].tolist()

# Parameters
batch_size = 1000
num_batches = (len(unmatched_df) // batch_size) + 1

results = []

for i in range(num_batches):
    batch_df = unmatched_df.iloc[i * batch_size : (i + 1) * batch_size]
    print(f"🔄 Processing batch {i + 1} of {num_batches}...")

    for _, row in batch_df.iterrows():
        mk = str(row['match_key'])

        match_result = process.extractOne(mk, valid_keys, score_cutoff=96)
        if match_result:
            match_key_found, score, _ = match_result
            song_match_row = songs_meta_df[songs_meta_df['match_key'] == match_key_found].iloc[0]
            print(f'Found: {match_result}')

            results.append({
                'original_match_key': mk,
                'original_title': row['Title'],
                'original_artist': row['Artist'],
                'fuzzy_match_key': match_key_found,
                'matched_song': song_match_row['song'],
                'matched_artist': song_match_row['Artists'],
                'matched_song_id': song_match_row['song_id'],
                'similarity_score': score
            })

# Final DataFrame
results_df = pd.DataFrame(results)
results_df.head(10)  # optional display
```

```python
# Create a copy and merge with fuzzy matches
hits_df_matched = hits_df_indexed.merge(
    results_df[['original_match_key', 'matched_song_id']],
    left_on='match_key',
    right_on='original_match_key',
    how='left'
)

# Fill missing song_id with matched_song_id from fuzzy results
hits_df_matched['song_id'] = hits_df_matched['song_id'].fillna(hits_df_matched['matched_song_id'])

# Drop helper columns
hits_df_matched.drop(columns=['original_match_key', 'matched_song_id'], inplace=True)
```

```python
hits_df_matched.head(10)
```

```python
# Count matched rows
num_matched = hits_df_matched['song_id'].notna().sum()

# Count total rows
total_rows = len(hits_df_matched)

# Calculate percentage
percentage_matched = (num_matched / total_rows) * 100

print(f"Matched songs: {num_matched} of {total_rows} ({percentage_matched:.2f}%)")
```

```python
# Step 0: Make a copy of the matched DataFrame
hits_df_final = hits_df_matched.copy()

# Step 1: Drop unnecessary columns (only if they exist)
cols_to_drop = ['Title', 'Artist', 'Label', 'Title_clean', 'Artist_clean', 'match_key']
hits_df_final.drop(columns=[col for col in cols_to_drop if col in hits_df_final.columns], inplace=True)

# Step 2: Convert Year and Week to Month using the first day of the week (Monday)
def week_to_month(year, week):
    try:
        week_start = datetime.strptime(f"{int(year)}-{int(week)}-1", "%Y-%W-%w")
        return week_start.month
    except:
        return None

hits_df_final['month'] = hits_df_final.apply(lambda row: week_to_month(row['Year'], row['Week']), axis=1)

# Step 3: Create weight column = 1 / Position, rounded to 3 decimals
hits_df_final['weight'] = (1 / hits_df_final['Position']).round(3)

# Step 4: Keep only rows with song_id that is not NaN
hits_df_final = hits_df_final[hits_df_final['song_id'].notna()].copy()

# Step 5: Convert song_id to integer
hits_df_final['song_id'] = hits_df_final['song_id'].astype(int)
```

```python
unique_song_count = hits_df_final['song_id'].nunique()
print(f"Number of unique songs: {unique_song_count}")
```

```python
# Group by title + artist and check if there is a mix of NaN and non-NaN in song_id
inconsistent = (
    hits_df_final
    .groupby(['Title', 'Artist'])['song_id']
    .apply(lambda x: x.isna().any() and x.notna().any())
)

# Filter only inconsistent songs
inconsistent = inconsistent[inconsistent]

# Show results (True = has rows with and without song_id)
print("Songs with inconsistently assigned song_id:")
print(inconsistent)
```
