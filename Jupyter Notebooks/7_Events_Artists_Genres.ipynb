```python
def categorize_genres(genres_list):
    genres_str = ','.join(genres_list).lower()
    if any(g in genres_str for g in ['rock', 'emo', 'alternative rock']):
        return 'rock'
    elif any(g in genres_str for g in ['hardcore', 'punk', 'grunge']):
        return 'punk'
    elif any(g in genres_str for g in ['metal', 'metalcore', 'deathcore', 'nu metal']):
        return 'metal'
    elif any(g in genres_str for g in ['pop', 'indie pop', 'synthpop']):
        return 'pop'
    elif any(g in genres_str for g in ['hip hop', 'hip-hop', 'rap', 'grime', 'trap']):
        return 'hip hop'
    elif any(g in genres_str for g in ['country', 'folk', 'bluegrass']):
        return 'country'
    else:
        # return 'other'
        return genres_str
```

```python
filtered_df['Genre_general'] = filtered_df['Genre_list'].apply(categorize_genres)
```
